```python
import matplotlib.pyplot as plt
import numpy as np



# Scatter plot
plt.scatter(merged_df['price_off_peak_var'], merged_df['cons_12m'], c=merged_df['churn'])
plt.xlabel('Price Off Peak Variance')
plt.ylabel('Consumption 12 Months')
plt.title('Scatter Plot: Price vs. Consumption')
plt.colorbar(label='Churn')
plt.show()



```

```python
# Box plot
merged_df.boxplot(column='cons_12m', by='churn')
plt.title('Box Plot: Consumption Distribution by Churn')
plt.ylabel('Consumption 12 Months')
plt.xlabel('Churn')
plt.show()
```

```python
# Bar plot
merged_df.groupby('churn')['cons_12m'].mean().plot(kind='bar')
plt.title('Bar Plot: Average Consumption by Churn')
plt.ylabel('Average Consumption 12 Months')
plt.xlabel('Churn')
plt.show()
```
